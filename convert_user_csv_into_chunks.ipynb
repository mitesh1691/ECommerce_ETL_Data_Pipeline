```python
import pandas as pd
```

## Creating Chunks 1 to 5 from 1st CSV file

```python
df = pd.read_csv('/content/6M-0K-99K.users.dataset.public.csv')
```

```python
df.head()
```

```python
df.count()
```

```python
len(df)
```

```python
chunk_size = len(df) // 5 + (1 if len(df) % 5 else 0)


for i in range(5):
  start = i * chunk_size
  end = start + chunk_size
  chunk_df = df.iloc[start:end]
  chunk_df.to_csv(f'/content/chunk_data1/chunk_{i+1}.csv', index=False)
```
